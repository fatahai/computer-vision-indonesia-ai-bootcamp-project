# **PEMROSESAN DATA DAN REKAYASA FITUR**

```python
# pisah fitur dan label pada data latih
fitur_latih = data_latih.drop(['SalePrice'], axis=1)
label_latih = data_latih['SalePrice']
```

```python
# penggabungan data latih dan data uji
dataset = pd.concat([fitur_latih, data_uji], axis=0)
```

```python
# informasi data yang sudah digabung (dataset)
dataset.info()
```

## Data yang hilang

```python
# menghitung data yang hilang
data_hilang_total = dataset.isnull().sum()
data_hilang_persen = dataset.isnull().sum() / len(dataset) * 100

# membuat tabel data yang hilang
data_hilang = pd.DataFrame({
    'Total': data_hilang_total,
    'Persentase': data_hilang_persen
})

data_hilang = data_hilang[data_hilang['Total'] > 0]
data_hilang.sort_values(by='Total', ascending=False, inplace=True)
data_hilang
```

```python
from sklearn.impute import SimpleImputer

# Group 1:
# fitur kategorik yang memiliki data kosong
group_1 = [
    'PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType',
    'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType'
]
dataset[group_1] = dataset[group_1].fillna("None")

# Group 2:
# fitur numerik yang memeiliki data kosong
group_2 = [
    'GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF',
    'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea'
]
dataset[group_2] = dataset[group_2].fillna(0)

# Group 3:
# fitur lainnya
group_3a = [
    'Functional', 'MSZoning', 'Electrical', 'KitchenQual', 'Exterior1st',
    'Exterior2nd', 'SaleType', 'Utilities'
]
imputer = SimpleImputer(strategy='most_frequent')
dataset[group_3a] = pd.DataFrame(imputer.fit_transform(dataset[group_3a]), index=dataset.index)

dataset.LotFrontage = dataset.LotFrontage.fillna(dataset.LotFrontage.mean())
dataset.GarageYrBlt = dataset.GarageYrBlt.fillna(dataset.YearBuilt)
```

```python
# periksa data yang hilang
sum(dataset.isnull().sum())
```

```python
# perlihatkan dataset yang sudah dibersihkan
dataset
```

## *Feature scaling*

```python
# pen-skala-an kolom numerik
from sklearn.preprocessing import RobustScaler
kolom_numerik = list(dataset.select_dtypes(exclude='object').columns)
scaler = RobustScaler()
dataset[kolom_numerik] = scaler.fit_transform(dataset[kolom_numerik])
```

```python
# perlihatkan dataset yang sudah di-skala-kan
dataset
```

## One-hot encoding

```python
# one-hot encoding bertujuan untuk mengubah data kategorik menjadi numerik
dataset_modif = pd.get_dummies(dataset, drop_first=True)
dataset_modif.info()
```

## Pisah data

```python
# jumlah data latih
# n_latih = len(data_latih)

# dataset latih
data_train = dataset_modif.iloc[:1460, :]

# dataset uji
data_test = dataset_modif.iloc[1460:, :]

print(f"Ukuran data latih : {data_train.shape}")
print(f"Ukuran data uji : {data_test.shape}")
```

```python
# data label
target_train = label_latih
target_train
```

# **PEMODELAN DATA**

```python
# import model random forest
from sklearn.ensemble import RandomForestRegressor
```

```python
# training model
random_forest_regresi = RandomForestRegressor(n_estimators=100, criterion='absolute_error', max_depth=10)
random_forest_regresi.fit(data_train, target_train)
```

```python
# metriks evaluasi performa training
model_score_train = random_forest_regresi.score(data_train, target_train)
print(f"Akurasi model training adalah {round(model_score_train, 4)}")
```
